```python
# Import necessary libraries
import cv2
import numpy as np
from IPython.display import display, clear_output
from PIL import Image

# Capture video/video path
cap = cv2.VideoCapture('D:\ML_Practice\Traffic_Management\cars.mp4')

# Use trained cars XML classifiers
car_cascade = cv2.CascadeClassifier('D:\ML_Practice\Traffic_Management\haarcascade_cars.xml')

# Read until video is completed
while True:
    # Capture frame by frame
    ret, frame = cap.read()
    if not ret:
        break  # Exit loop if no frames are returned or video has ended

    # Convert video into gray scale of each frame
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    # Detect cars in the video
    cars = car_cascade.detectMultiScale(gray, 1.1, 3)

    # Draw a rectangle around each car
    for (x, y, w, h) in cars:
        cv2.rectangle(frame, (x, y), (x+w, y+h), (0, 255, 0), 2)

    # Convert frame to PIL Image for display in Jupyter
    rgb_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
    pil_img = Image.fromarray(rgb_frame)

    # Display the image in Jupyter Notebook
    clear_output(wait=True)
    display(pil_img)

    # Optionally add a small delay for video playback
    cv2.waitKey(30)

# Release the video-capture object
cap.release()
```
